```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

Länge=np.array([165,114,120,122,132,120,210,132,148,118,156,120,121,178,120,127,180,142.5,140,121,122,152,130,150,100,120,120,150,150,150,150,120,150,150,120,120,115,132,114,120,160,130,100])
Breite=np.array([112,100,89,104,100,100,100,100,118,100,110,100,91,120,100,94,106,106,89,91,91,122,100,130,100,110,110,100,120,100,120,100,120,100,100,100,100,98,100,100,120,100,100])
Dicke=np.array([12.5,10,10,9,10,9,10,14.5,12.5,10,12,10,10,10,9,12,12,13,10,12.5,12.5,12.5,14,12,9,10,10,14,15,14,14,11,12,11,10,10,10,12,10,10,14,11,9])
Gewicht=np.array([9.25,4.33,3.775,4.75,5.8,4.4,8.3,6.2,9.4,5.7,9.1,5.3,3.35,8.48,4.5,6.2,10.9,7.5,8.16,6.2,6.35,8.6,5.8,6,3.2,4.9,4.9,8.4,9.6,8.1,9.5,8.6,8.6,7.5,3.9,3.9,6,6.3,3.8,6,8,6,5])
Preis=np.array([339,169,126,211,259,179,399,279,351,269,253,167,159,369,179,169,269,249,199,161,194,264,173,474,195,258,218,349,399,349,399,229,349,249,149,169,219,339,165,279,361,237,160])
Volumen=np.zeros(len(Länge))
Nummern=np.zeros(len(Länge))
Namen=np.array(['BD Mondo', 'BD Impact', 'BD Circuit', 'BD Dropzone', 'Ocun Moonwalk', 'Ocun Sundance', 'Ocun Incubator', 'Ocun Dominator', 'Petzl Cirro', 'Petzl Alto','Mammut Crashiano','Mammut Slam','Metolius Bergfr. Basic','Metolius Bergfr. Magnum','Ocun  Bergfr. Kava V2','MadRock Mad Pad','MadRock Triple Mad Pad','MadRock Duo','MadRock R3','MadRock Hera','Evolv Drop Pad','Evolv Launch Pad','Beal Double AirBag','Beal Jumbo','Beal AirLight','Grivel Trende','Grivel Grivel','Snap Wham','Snap Grand Wham',' Snap Guts','Snap Grand Guts','Snap Rebound','Snap Grand Rebound','Snap Grand Stamina','Snap Hip','Snap Hop','La Sportiva Laspo Pad','Cassin Domino','Cassin Minido','EB Connect Pad','Moon Saturn','Moon Warrior','Moon Pluto'])
Dummy=np.zeros(len(Länge))
#Volumen-array erstellen

for i in range(0,len(Länge)):
    Volumen[i]=Länge[i]*Breite[i]*Dicke[i]*(1/1000)
    Nummern[i]=i
    Dummy[i]=0.05

#VPG

VPG=Volumen/Preis
Minimum=min(VPG)
Maximum=max(VPG)
Delta=Maximum-Minimum


#Farben array erstellen
Farben=np.zeros(len(VPG))
for i in range(0,len(VPG)):
    Farben[i]=100*(VPG[i]-Minimum)/Delta

#Plots

figure(figsize=(25,10))
plt.plot(Nummern,Volumen/Preis,label='Volumen/Geld [Liter/Euro]',linewidth=4)
plt.plot(Nummern,20*Gewicht/Preis,label='Gewicht/Geld [*50 g/Euro]')
plt.plot(Nummern,10*Gewicht/Volumen,label='Dichte [*0.1 g/m^3]')
plt.plot(Nummern,Volumen/1000,label='Volumen [*10^3 Liter]')
plt.plot(Nummern,Preis/1000,label='Preis [*10^3 Euro]')
plt.scatter(Nummern,Dummy , c=Farben , s=300 , cmap='RdYlGn',label='Grün => gute Preis-Leistung')

plt.grid()
plt.legend()

locs, labels = plt.xticks()  # Get the current locations and labels.
plt.xticks(np.arange(0, len(Länge), step=1))  # Set label locations.
plt.xticks(np.arange(len(Länge)), Namen , rotation=70)  # Set text labels.
#xticks([0, 1, 2], ['January', 'February', 'March'],rotation=20)  # Set text labels and properties.
#plt.xticks([])  # Disable xticks.

plt.savefig("Crashpads.jpg")
```
